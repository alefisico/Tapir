```python
def get_lr_sig_bg(hn):
	h1 = asrootpy(tf.Get(hn).Clone())

	h_bb = asrootpy(h1.ProjectionX("bb", 1, 1))
	h_bj = asrootpy(h1.ProjectionX("bj", 2, 2))
	h_cc = asrootpy(h1.ProjectionX("cc", 3, 3))
	h_jj = asrootpy(h1.ProjectionX("jj", 4, 4))
	h_u = asrootpy(h1.ProjectionX("u", 5, 5))

	h_sig = h_bb.Clone("sig")
	h_bg = h_jj.Clone("bg")
	h_bg.Add(h_bj)
	h_bg.Add(h_cc)
	
	he.normalize(h_bg)
	he.normalize(h_sig)

	return h_bg, h_sig

def get_cumulative(h_sig, h_bg):
	h_sig_cum = asrootpy(h_sig.GetCumulative())
	h_bg_cum = asrootpy(h_bg.GetCumulative())

	h_sig_cum.Scale(1.0 / h_sig.Integral())
	h_bg_cum.Scale(1.0 / h_bg.Integral())

	return (h_bg_cum, h_sig_cum)


def roc(bg, sig):
	assert(bg.size == sig.size)

	bg = bg / bg[-1]
	sig = sig / sig[-1]

	assert(bg[-1] == 1)
	assert(sig[-1] == 1)

	out = numpy.zeros((bg.size, 2))
	for i in range(bg.size):
		out[i, 0] = sig[i]
		out[i, 1] = bg[i]
	return out

def get_lr_sig_bg_2(hn):
    h1 = asrootpy(tf.Get("ttjets_13TeV_" + hn).Clone())
    h2 = asrootpy(tf.Get("tthbb_13TeV_" + hn).Clone())
    h1 = asrootpy(h1.projection_x())
    h2 = asrootpy(h2.projection_x())
    return h1, h2
```

```python
hlr1 = get_lr_sig_bg("ttjets_13TeV_btag_lr_rad")
hlr2 = get_lr_sig_bg("ttjets_13TeV_btag_lr2_rad")
hlr3 = get_lr_sig_bg("ttjets_13TeV_btag_lr3_rad")
hlr4 = get_lr_sig_bg("ttjets_13TeV_btag_lr4_rad")

hlr5 = get_lr_sig_bg_2("btag_lr2_rad")
hlr6 = get_lr_sig_bg_2("btag_lr3_rad")
hlr7 = get_lr_sig_bg_2("btag_lr4_rad")
```

```python
lr1 = get_cumulative(*hlr1)
lr2 = get_cumulative(*hlr2)
lr3 = get_cumulative(*hlr3)
lr4 = get_cumulative(*hlr4)
lr5 = get_cumulative(*hlr5)
lr6 = get_cumulative(*hlr6)
lr7 = get_cumulative(*hlr7)
```

```python
roc_lr1 = roc(he.as_array(lr1[0]), he.as_array(lr1[1]))
roc_lr2 = roc(he.as_array(lr2[0]), he.as_array(lr2[1]))
roc_lr3 = roc(he.as_array(lr3[0]), he.as_array(lr3[1]))
roc_lr4 = roc(he.as_array(lr4[0]), he.as_array(lr4[1]))
roc_lr5 = roc(he.as_array(lr5[0]), he.as_array(lr5[1]))
roc_lr6 = roc(he.as_array(lr6[0]), he.as_array(lr6[1]))
roc_lr7 = roc(he.as_array(lr7[0]), he.as_array(lr7[1]))
```

```python
fig = plt.figure(figsize=(8,8))
ax = plt.axes()
plt.plot(roc_lr1[:, 0], roc_lr1[:, 1], label="default")
plt.plot(roc_lr2[:, 0], roc_lr2[:, 1], label="default (2)", marker=".", lw=0)
plt.plot(roc_lr3[:, 0], roc_lr3[:, 1], label="W cq", marker="x")
plt.plot(roc_lr4[:, 0], roc_lr4[:, 1], label="cc radiation", marker="o")
plt.legend(loc=2)
plt.xlim(0, 0.2)
plt.ylim(0.0, 0.1)
plt.ylabel("bj, cc, jj acceptance")
plt.xlabel("bb acceptance")
ax
```

```python

def get_lh_hists(l_bbbb, l_bbcc, l_bbjj):
    
    hs = []
    for s in ["ttjets", "tthbb"]:
        tf1 = ROOT.TFile("/Users/joosep/Documents/tth/data/{0}.root".format(s))
        tt1 = tf1.Get("tree")
        ROOT.gROOT.cd()
        tt1.Draw(
            "btag_lr_l_bbbb / ({0}*btag_lr_l_bbbb + {1}*btag_lr_l_bbcc + {2}*btag_lr_l_bbjj) >> h(500, 0, 1)".format(l_bbbb, l_bbcc, l_bbjj),
            "weight * btag_lr_l_bbbb > 0.0"
        )
        h = asrootpy(ROOT.gROOT.Get("h").Clone(s))
        he.normalize(h)
        hs += [h]
    return hs
```

```python
def get_roc(*args):
    h1, h2 = get_cumulative(*get_lh_hists(*args))
    r = roc(he.as_array(h1), he.as_array(h2))
    return r
```

```python
r1 = get_roc(1.0, 0.0, 1.0)
r2 = get_roc(1.0, 1.0, 1.0)
r3 = get_roc(1.0, 0.022, 0.015)
```

```python
fig = plt.figure(figsize=(8,8))
ax = plt.axes()
plt.plot(r1[:, 0], r1[:, 1], lw=0, marker="x")
plt.plot(r2[:, 0], r2[:, 1], lw=0, marker="o")
plt.plot(r3[:, 0], r3[:, 1], lw=0, marker="^")
plt.xlim(0.0, 0.2)
plt.ylim(0.0, 0.05)
```
